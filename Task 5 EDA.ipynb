```python
cat_cols=data.select_dtypes(include=['object']).columns
num_cols = data.select_dtypes(include=np.number).columns.tolist()
print("Categorical Variables:")
print(cat_cols)
print("Numerical Variables:")
print(num_cols)
```

 #Iterating through a list of numerical column names (num_cols) in a DataFrame, and for each column, you are printing its skewness, plotting a histogram, and a boxplot using the matplotlib and seaborn libraries. The skewness is a measure of the asymmetry of the probability distribution of a real-valued random variable about its mean.#

```python
for col in num_cols:
    print(col)
    print('Skew :', round(data[col].skew(), 2))
    plt.figure(figsize = (15, 4))
    plt.subplot(1, 2, 1)
    data[col].hist(grid=False)
    plt.ylabel('count')
    plt.subplot(1, 2, 2)
    sns.boxplot(x=data[col])
    plt.show()
```

```python
plt.figure(figsize=(13,17))
sns.pairplot(data=data.drop(['Driven (Kms)','Price'],axis=1))
plt.show()
```
